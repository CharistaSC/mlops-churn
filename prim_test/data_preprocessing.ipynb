### üî¢ One-Hot Encoding: What It Is and Why We Need To Use It

#### üìå What is One-Hot Encoding?

One-Hot Encoding is a technique used to convert **categorical variables** into a format that can be provided to **machine learning models**, which typically require **numerical input**.

Instead of assigning arbitrary numeric values to categories (which can imply false ordinal relationships), One-Hot Encoding creates **new binary columns** ‚Äî one for each unique category ‚Äî where:
- A value of `1` indicates the presence of that category.
- A value of `0` indicates its absence.

---

#### üö´ Why Not Just Use Numbers like 0, 1, 2?

Assigning numbers directly (e.g., `Contract: Month-to-month = 0, One year = 1, Two year = 2`) implies a **ranking or distance** between categories. This can mislead the model into thinking, for example, that `Two year` is "twice as long" or "better" than `One year`, which is not the case unless it's a truly ordinal feature.

---

#### ‚úÖ Why One-Hot Encoding is Needed in This Project

In the Telco Customer Churn dataset, the following columns contain **nominal categorical data** (unordered categories), which must be one-hot encoded for accurate model training:

| Column           | Reason for Encoding |
|------------------|---------------------|
| `Contract`        | Represents plan types `Month-to-month`, `One year`, `Two year` have no natural order in terms of churn risk |
| `PaymentMethod`   | Categories like `Credit card`, `Electronic check`, `Bank transfer`, `Mailed check` are distinct, with no inherent ranking |
| `InternetService` | Options like `DSL`, `Fiber optic`, and `No` should be treated as separate service types |

By converting these into one-hot encoded vectors, we:
- Enable the model to understand each category independently.
- Avoid introducing misleading assumptions about category relationships.
- Prepare the data in a structure suitable for most supervised ML models.

---

```python
# Columns to one-hot encode
one_hot_cols = ['Contract', 'PaymentMethod', 'InternetService']

# Get original column order
original_cols = df.columns.tolist()

# Perform one-hot encoding
df_encoded = pd.get_dummies(df[one_hot_cols], prefix=one_hot_cols, drop_first=False)

# Drop original categorical columns from df
df = df.drop(columns=one_hot_cols)

# Prepare a list to hold final column order
new_cols = []

# Rebuild column order: loop through original columns
for col in original_cols:
    if col in one_hot_cols:
        # Insert all the new one-hot columns for this original column
        new_cols.extend([c for c in df_encoded.columns if c.startswith(col + '_')])
    elif col in df.columns:
        # Retain non-encoded original columns
        new_cols.append(col)

# Combine df with encoded columns (in original layout)
df = pd.concat([df, df_encoded], axis=1)
df = df[new_cols]  # Reorder to match desired layout

# Convert all boolean columns to integers (0/1)
bool_cols = df.select_dtypes(include=['bool']).columns
df[bool_cols] = df[bool_cols].astype(int)

# Convert any Yes and No to 1 and 0
df.replace({'Yes': 1, 'No': 0}, inplace=True)
```

### ‚öñÔ∏è Feature Scaling in Machine Learning

#### üìå What is Feature Scaling?

Feature scaling is the process of transforming numerical values in your dataset so they fall within a similar range or distribution. This helps machine learning models **learn more effectively** and **perform more accurately** ‚Äî especially when features vary in magnitude, units, or importance.

---

#### ‚ùì Why Do We Scale Features?

Machine learning models **rely on numerical calculations** such as distance (KNN, SVM), gradient descent (Logistic Regression, Neural Networks), and regularisation (Ridge, Lasso). If some features have significantly larger values than others, they can:

- **Dominate the learning process**
- **Slow convergence** in gradient-based models
- **Skew model interpretation** or bias feature importance

---

#### üî¢ Common Types of Scalers

##### 1. **Min-Max Scaler (Normalization)**

- Scales values to a fixed range, usually [0, 1]
- Preserves the shape of the original distribution
- ‚ö†Ô∏è Sensitive to outliers

üìò Formula:
$$X_{\text{scaled}} = \frac{X - X_{\min}}{X_{\max} - X_{\min}}$$

üìå Use When:
- You don‚Äôt have outliers
- You need all features to be on the same scale (e.g., Neural Networks)

---

##### 2. **Standard Scaler (Z-score Standardization)**

- Centers data to mean = 0 and standard deviation = 1
- Allows for negative values
- Assumes Gaussian-like distribution (but not strictly required)

üìò Formula:
$$X_{\text{scaled}} = \frac{X - \mu}{\sigma}$$

üìå Use When:
- Your features are normally distributed or roughly symmetrical
- You're using models like Logistic Regression, SVM, or PCA

---

##### 3. **Robust Scaler**

- Uses **median and interquartile range (IQR)** instead of mean and standard deviation
- Resistant to outliers
- Does not strictly bound the data between 0 and 1

üìò Formula:
$$X_{\text{scaled}} = \frac{X - \text{median}}{\text{IQR}}$$

üìå Use When:
- Your data has **extreme values or outliers**

---

#### üß† When is Scaling Not Needed?

You can skip scaling if:
- You're using **tree-based models** (e.g., Decision Trees, Random Forest, XGBoost), which are not sensitive to feature magnitude.
- All your features are already on a similar scale.

---

#### ‚úÖ Summary: Choosing the Right Scaler

| Scaler           | Use Case                                        |
|------------------|-------------------------------------------------|
| MinMaxScaler     | Neural Networks, KNN, features with no outliers |
| StandardScaler   | Linear models, PCA, SVM                         |
| RobustScaler     | Outlier-heavy datasets                          |

Always apply scaling **after splitting** categorical and numerical features, and **before training your model**.

## üìä Segmenting Users by Service Combinations

We can divide customers into **three distinct subsets** based on their subscription to Phone and Internet services:

---

###### 1. Users with **Both Phone and Internet**
- üí° **Most engaged** user group
- Access to both service types: phone and internet
- All feature columns are relevant and applicable
- Expect meaningful variation across:
  - OnlineSecurity
  - TechSupport
  - Streaming services
  - MultipleLines

---

###### 2. Users with **Only Phone Service**
- `InternetService` = `"No"`
- All internet-related features contain `"No internet service"` and can be **dropped**:
  - OnlineSecurity
  - OnlineBackup
  - DeviceProtection
  - TechSupport
  - StreamingTV
  - StreamingMovies
- Focus is on voice service usage
- May represent users with simpler needs or limited infrastructure

---

###### 3. Users with **Only Internet Service**
- `PhoneService` = `"No"`
- `MultipleLines` = `"No phone service"` ‚Üí can be **dropped**
- Retains all internet-related features
- Likely more **tech-savvy** or digitally-oriented users
- Streamlined feature set with fewer voice-service variables

---

By segmenting users this way, we can:
- Train specialised models per group
- Simplify feature cleaning and encoding
- Discover behaviour patterns unique to each user type


```python
# Take note, InternetService_No is using 0: False and 1:True, so the logic is inverted

# Group 1: Users with both Phone and Internet services
both_services_df = df[(df['PhoneService'] == 1) & (df['InternetService_No'] == 0)].copy()

# Group 2: Users with only Phone service
only_phone_df = df[(df['PhoneService'] == 1) & (df['InternetService_No'] == 1)].copy()

# Group 3: Users with only Internet service
only_internet_df = df[(df['PhoneService'] == 0) & (df['InternetService_No'] == 0)].copy()

# Drop PhoneService column since the only factor in difference is the type of internet service
both_services_df = both_services_df.drop(columns=[
    'PhoneService',
    'InternetService_No'
])

# Drop all internet-related and PhoneService columns.
only_phone_df = only_phone_df.drop(columns=[
    'PhoneService',
    'InternetService_DSL',
    'InternetService_Fiber optic',
    'InternetService_No',
    'OnlineSecurity',
    'OnlineBackup',
    'DeviceProtection',
    'TechSupport',
    'StreamingTV',
    'StreamingMovies'
])

# Drop all phone-related and InternetService_No columns.
only_internet_df = only_internet_df.drop(columns=[
    'PhoneService',
    'InternetService_No',
    'MultipleLines'
])
```

### ü§ñ Supervised Machine Learning Models for Churn Prediction

In churn prediction, we know the target variable (`Churn`: Yes/No), which makes this a **supervised learning problem**. Below are commonly used supervised ML models, how they work, and their strengths and limitations in the context of predicting customer churn.
---
#### 1. üî¢ Logistic Regression
##### üõ†Ô∏è How It Works:
- A linear model that predicts the probability of an event (churn) using a logistic (sigmoid) function.
- Outputs values between 0 and 1 (interpreted as probabilities).
##### ‚úÖ Pros:
- Fast to train and deploy.
- Easy to interpret (good for business stakeholders).
- Works well with binary classification.
##### ‚ùå Cons:
- Assumes a linear relationship between features and the log-odds of churn.
- Underperforms on complex or nonlinear data.
- Sensitive to multicollinearity and outliers.
**üîé Best for:** Baseline model, feature understanding, and interpretable results.
---
#### 2. üå≤ Random Forest
##### üõ†Ô∏è How It Works:
- An ensemble of decision trees trained on bootstrapped samples.
- Each tree votes; the majority vote is the final prediction.
##### ‚úÖ Pros:
- Handles both categorical and numerical data.
- Resistant to overfitting compared to individual trees.
- Provides feature importance scores.
##### ‚ùå Cons:
- Slower to train and less interpretable than logistic regression.
- Model size can be large (less ideal for lightweight deployments).
**üîé Best for:** Balanced accuracy and robustness with little tuning.
---
#### 3. ‚ö° XGBoost / LightGBM
##### üõ†Ô∏è How It Works:
- Gradient boosting algorithms that build trees sequentially to minimize error.
- XGBoost focuses on speed and performance; LightGBM improves memory usage and scalability.
##### ‚úÖ Pros:
- High accuracy and fast performance.
- Handles missing values and categorical features well.
- Works well on tabular data like churn datasets.
##### ‚ùå Cons:
- Requires careful tuning (learning rate, depth, etc.).
- Less interpretable unless you use tools like SHAP or LIME.
**üîé Best for:** Production-level performance with high accuracy demands.
---
#### 4. ü§ù K-Nearest Neighbours (KNN)
##### üõ†Ô∏è How It Works:
- Classifies a data point by majority vote of its K closest neighbors in feature space.
##### ‚úÖ Pros:
- Intuitive and simple.
- No training phase (lazy learning).
##### ‚ùå Cons:
- Very slow on large datasets (especially during inference).
- Requires scaling and good distance metrics.
- Performance drops with high-dimensional data.
**üîé Best for:** Simple baselines, small datasets.
---
#### 5. üìê Support Vector Machine (SVM)
##### üõ†Ô∏è How It Works:
- Finds the hyperplane that best separates the data into classes with the maximum margin.
##### ‚úÖ Pros:
- Effective on small datasets with clear boundaries.
- Works well in high-dimensional spaces.
##### ‚ùå Cons:
- Slow with large datasets.
- Not great with overlapping or noisy classes.
- Requires scaling and parameter tuning.
**üîé Best for:** Clean, medium-sized datasets with well-separated classes.
---
#### 6. üß† Neural Networks
##### üõ†Ô∏è How It Works:
- Layers of artificial neurons learn feature interactions through backpropagation.
- Can model complex, non-linear relationships.
##### ‚úÖ Pros:
- Very powerful for large datasets and feature-rich inputs.
- Can learn intricate patterns in data.
##### ‚ùå Cons:
- Requires more data and tuning.
- Black-box ‚Äî hard to explain.
- Prone to overfitting if not regularized.
**üîé Best for:** Experimentation or integration into deep-learning pipelines.
